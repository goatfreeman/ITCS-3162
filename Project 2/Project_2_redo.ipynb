```python
cols = list(loan.columns)
cols.remove('loan_status')
loan = pd.get_dummies(loan, columns=cols)
```

```python
loan
```

```python
X = loan.drop(columns = ['loan_status'])
y = loan['loan_status']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (48,34), dpi=300)
tree.plot_tree(clf, max_depth = None, feature_names = X.columns, filled=True)
plt.show()
```

```python
from sklearn import metrics
```

```python
predicted = clf.predict(X_test)
```

```python
print(metrics.classification_report(y_test, predicted))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, predicted)
```

```python
cm = confusion_matrix(y_test, predicted)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for Loan Approval")
plt.show()
```
